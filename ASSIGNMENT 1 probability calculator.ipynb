```python
                                              #PROBABILITY CALCULATOR
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

#Taking user input for cdf
n1=int(input("ENTER THE FIRST NUMBER [mean]:"))
n2=int(input("ENTER THE SECOND NUMBER[sigma]:"))
n3=int(input("ENTER THE NUMBER FOR WHICH YOU HAVE TO FIND CDF:"))
from statistics import NormalDist
number = NormalDist(mu=n1,sigma=n2)
cdf=number.cdf(n3)
print(f"The CDF is {cdf}")
#Calculating PDF
N = 500
  
# initializing random values for pdf
data = np.random.randn(N)
  
count, bins_count = np.histogram(data, bins=10)
  
# finding the PDF of the histogram using count values
pdf = count / sum(count)
print(f"The PDF is {pdf}")

cdf1 = np.cumsum(pdf)

plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf1, label="CDF")
plt.legend()



```
