```python
name=input("Enter the student name:")
dept=input("Enter the department:")
year=input("Enter the year of the student:")
sem=input("Enter the semester:")
n=int(input("Enter the number of subjects:"))
print("Grade details:")
print("O:10\nA+:9.0\nA:8.0\nB+:7.0\nB:6.0\nC:5.0\nU:fail")
d={'O':'10','A+':'9','A':'8','B+':'7','B':'6','C':'5','U':'fail'}
grade=[]
try:
    for i in range(n):
        sub=input("Enter the subject code:")
        gd=input("Enter the grade:")
        grade.append(int(d[gd]))
    print("******************************************************************************************")
    print("STUDENT DETAILS:\n")
    print("NAME:{}\t\tYEAR:{}\n".format(name,year))
    print("DEPARTMENT:{}\t\tSEMESTER:{}".format(dept,sem))
    print("******************************************************************************************")
    print("THE CGPA:{:.2f}".format(((sum(grade)/n)*9.5)/10))
    print("******************************************************************************************")
except:
    print("THE CGPA:ARREAR(FAIL)")
      
```
